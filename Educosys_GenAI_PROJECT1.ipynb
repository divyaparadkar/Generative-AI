```python
#Accept a no and check if it a perfect no  or not
n=int(input("check ur no is perfect or not "))
sum=0
for i in range(1,n):
  if n%i ==0:
    sum=sum+i
if sum==n:
  print("perfect no")
else:
  print("not perfect no")
```

```python
#check weather the no is prime or not
n=int(input("enter ur no is prime or not "))
count =0
for i in range(1,n+1):
  if n%i==0:
    count=count+1
if count ==2:
  print("your no is prime")

else:
  print("not prime")
```
